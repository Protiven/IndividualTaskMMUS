## Initial

```python
import numpy as np
import scipy as sp
from matplotlib import pyplot as plt
from numpy.linalg import inv
import matplotlib as mplb
from matplotlib.font_manager import FontProperties
%matplotlib inline
from numpy.random import normal
```

## Model

```python
import math as mth

N = 500
t = 0.5

# Настройка системы
F = lambda x: np.matrix([[x[0] + x[1] * t],
                         [x[1]]])

dF = lambda x: np.matrix([[1, t],
                          [0, 1]])

H = lambda x: np.array([[x[0]]])

dH = lambda x: np.array([[1, 0]])

# Настройка ошибок
sigma_a = 7100
sigma_n = 10

# Начальные условия (надо изменить)
X_0 = np.array([[0],
                [2]])

P_0 = np.matrix([[10, 0], 
                 [0, 10]])

# Ошибка системы (надо изменить)
G = np.array([[0], [2]])
Q = G.dot(G.transpose())*sigma_a**2

# Ошибка сенсора(надо изменить)
R = np.array([[sigma_n**2]])
```

```python
# Генерация измерений
# ideal_sign - идеальный сигнал
# y - сигнал измерителя/сенсора
t_ = 0
y = np.empty(N)
ideal_sign = np.empty(N)

y[0] = H(X_0) 
ideal_sign[0] = H(X_0) 
x_i = X_0
x_ist = X_0

for i in range(1, N):
    t_ = t*i
    
    # Для построения настоящих значений сенсора
    x_k_1 = np.array(F(x_i.T[0])) + np.array(G) * np.random.normal(0, Q)
    y[i] = H(np.array(F(x_i.T[0]))) + np.random.normal(0, R)
    x_i = x_k_1
    
    
    # Для построения истинных значений сенсора
    x_istina = np.array(F(x_ist.T[0]))
    ideal_sign[i] = H(np.array(F(x_ist.T[0])))
    x_ist = x_istina
```

## EKF

```python
def EKF(X_0, P_0, y, F, dF, H, dH, R, Q):
    X = np.zeros((len(y),*(X_0.shape)))
    P = np.zeros((len(y),*(P_0.shape)))
    K = np.zeros_like(X) 
    for i, _ in enumerate(y):
        #Prediction
        X[i] = F((X[i-1].T[0] if i > 0 else X_0.T[0]))
        P[i] = (F((X[i-1].T[0] if i > 0 else X_0.T[0])).T).dot((P[i-1] if i > 0 else P_0).dot(F((X[i-1].T[0] if i > 0 else X_0.T[0])))) + Q # Must change!!!!
        
        #Filtration
        K[i] = P[i].dot(dH(X[i]).T.dot(inv(dH(X[i]).dot(P[i].dot(dH(X[i]).T)) + R)))
        X[i] = X[i] + K[i] * (y[i] - H(X[i]))
        P[i] = (np.identity(X_0.shape[0]) - K[i].dot(dH(X[i]))).dot(P[i])
    return X, K

# Use Kalman
X, K = EKF(X_0, P_0, y, F, dF, H, dH, R, Q)
```

## Result

```python
fig, ax = plt.subplots(1, 1, figsize=(6,4), dpi = 500)
ax.plot(range(N), y, 'r', label = '$signal_{real}$', linewidth = 0.5 )
ax.plot(range(N), ideal_sign, 'g', label = '$signal_{ideal}$', linewidth = 1.2 )
ax.plot(range(N), X[:,0].transpose()[0],'b', label = '$signal_{EKF}$', linewidth = 1 )
ax.grid()
ax.legend(loc='upper left')
#ax.set_xlabel('points', fontsize = 7)
#ax.set_ylabel('magnitude', fontsize = 7);
```
